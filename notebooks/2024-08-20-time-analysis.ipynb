```python
import pandas as pd

df = (
    pd.read_csv("../custom_invlist_cpp/bench-results.csv")
    .groupby(['method', 'nb', 'nprobe', 'num_queries', 'k'])['dt']
    .median()
    .unstack('method')
    .assign(ratio=lambda df: df.comp/df.ref)
    ['ratio']
    .reset_index()
)

df['ratio'].describe()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df)
```
